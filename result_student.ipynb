```python
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
operator = input("Enter the operator (+, -, *, /): ")

if operator == "+":
    result = num1 + num2
elif operator == "-":
    result = num1 - num2
elif operator == "*":
    result = num1 * num2
elif operator == "/":
    if num2 != 0:
        result = num1 / num2
    else:
        print("Error: Division by zero is not allowed.")
        result = None
else:
    print("Error: Invalid operator.")
    result = None
    
    if result is not None:
    print("Result:", result)

```

```python
def create_student_dictionary(name, age, rollno, school):
    student_dict = {
        "name": name,
        "age": age,
        "rollno": rollno,
        "school": school
    }
    return student_dict


num_students = int(input("Enter the number of students: "))

students = [] 


for i in range(num_students):
    print(f"\nStudent {i+1}:")
    name = input("Enter student's name: ")
    age = int(input("Enter student's age: "))
    rollno = input("Enter student's rollno: ")
    school = input("Enter student's school: ")
    
    english = int(input("enter the english marks: "))
    maths = int(input("enter the marks: "))
    science=int(input("enter the marks: "))
    hindi=int(input("enter the marks: "))
    pysical=int(input("enter the marks: "))
    drawing=int(input("enter the marks: "))
    sanskit=int(input("enter the marks: "))


    student_details = create_student_dictionary(name, age, rollno, school,english,maths,science,hindi,pysical,drawing,sanskit)
    students.append(student_details)


api_data = {
    "data": students
}

print(api_data)

for i in  api_data["data"] :
    totalmarks=(english+maths+science+hindi+pysical+drawing+sanskit)
    print(i)
    mark=(totalmarks/7)
    print(mark)

    if  mark >= 90:
        grade = "A"
    elif mark >= 80:
        grade = "B"
    elif mark >= 70:
        grade = "C"
    elif mark >= 60:
        grade = "D"
    elif mark>= 50:
        grade = "F"
    elif mark>= 40:
        grade="E"
    elif mark>= 30:
        grade="g"
    else:
        grade="fail"
        print("Grade:", grade)

    print("\nMarks Sheet:")




    print("\n\n\t\t----------- Mark Sheet -----------\n\n")

    print("\t\tschool:",i["school"])
    print("\n\t\tfristname:",i["name"],"\t\tlast name:",i["name"])
    print("\n\t\trollno:",i["rollno"])

    marks=(round(mark,1))

    print(f"""  
---------------------------------------------- Mark Sheet -------------------------------------
|Subject       |\t\t\tMarks\t\t\t|\t\ttotal\t\t|
-----------------------------------------------------------------------------------------------
|English       |\t\t\t{english}\t\t\t|\t\t100\t\t|
|Maths         |\t\t\t{maths}\t\t\t|\t\t100\t\t|
|Science       |\t\t\t{science}\t\t\t|\t\t100\t\t|
|hindi         |\t\t\t{hindi}\t\t\t|\t\t100\t\t|
|pysical       |\t\t\t{pysical}\t\t\t|\t\t100\t\t|
|drawing       |\t\t\t{drawing}\t\t\t|\t\t100\t\t|
|sanskit       |\t\t\t{sanskit}\t\t\t|\t\t100\t\t|
----------------------------------------------------------------------------------------------
|Total Marks   |\t\t\t{totalmarks}\t\t\t|\t\t700\t\t|
|Average Marks |\t\t\t{marks}%\t\t\t|\t\t100\t\t|
|Grade         |\t\t\t{grade}\t\t\t|
----------------------------------------------------------------------------------------------
""")



    print(f"Total Marks: ", i["totalmarks"],"\t\tAverage Marks: ",i["marks"],"%")

    print("Grade: ", i["grade"])


    print(f"""
---------------------------grading system--------------------------
     100-90 | 90-80 | 80-70 | 70-60 | 60-50 |50-40 | 40-30 | 30-0 
        A   |    B  |   C   |     D |    E  |  F   |    F  |  FAIL
     """)

    print(f"""
             ---------------                          -----------------                 -----------------
             class reacher                                 principal                          parents
""")

    
           

```
